```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE as skTSNE
from methods.tSNE import tSNE as myTSNE
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
iris = load_iris()
X = iris.data
y = iris.target

my_tsne = myTSNE(perplexity=30, lr=100, GD_iters=500)
my_result = my_tsne.fit_transform(X)

sklearn_result = skTSNE(perplexity=30, learning_rate=100, n_iter=1000).fit_transform(X)

plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.scatter(my_result[:, 0], my_result[:, 1], c=y, cmap='viridis')
plt.title("My t-SNE")
plt.colorbar()

plt.subplot(122)
plt.scatter(sklearn_result[:, 0], sklearn_result[:, 1], c=y, cmap='viridis')
plt.title("Sklearn t-SNE")
plt.colorbar()

plt.tight_layout()
plt.show()
```
